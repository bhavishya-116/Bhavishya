```python
from nltk.tokenize import word_tokenize
text="Artificial Intelligence is transforming industries. NLP is a key part of AI. Tokenization is the first NLP step."
word_tokenize(text)
```

```python
from nltk.tokenize import sent_tokenize
text="Artificial Intelligence is transforming industries. NLP is a key part of AI. Tokenization is the first NLP step."
sent_tokenize(text)
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import PorterStemmer
stemmer= PorterStemmer()
words=["playing", "played", "plays", "happily", "studies"]
stemmed_words=[]
for word in words:
  stemmed_words.append(stemmer.stem(word))
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words =["running", "cars", "better", "children", "feet"]
lemmatized_words=[]
for word in words:
  lemmatized_words.append(lemmatizer.lemmatize(word))
print(lemmatized_words)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words=['talking','charging','filling','saw','said']

for i in words:
  s=stemmer.stem(i)
  l=lemmatizer.lemmatize(i)

  print(f"word: {i} |stem:{s}  |lemma:{l}\n")
```
